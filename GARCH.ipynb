```python
import yfinance as yf
import pandas as pd
import numpy as np
from arch import arch_model
import datetime as dt
import sys
import matplotlib.pyplot as plt
```

```python
ticker = "NVDA"

startdate = "2020-01-01"
enddate = "2024-07-30"

dfromyf = yf.download(ticker, startdate, enddate)
```

```python
print(dfromyf.head())
```

```python
dfromyf["log_return"] = dfromyf["Adj Close"].apply(lambda x: np.log(x)) - dfromyf["Adj Close"].shift(1).apply(lambda x: np.log(x))
```

```python
dataFin = dfromyf.dropna()
```

```python
dataFin.head()
```

```python
returns = dataFin["log_return"]
```

```python
model = arch_model(returns, mean="constant", vol='Garch', p=1, o=0, q=1, dist="Normal")

results = model.fit()

print(results.summary())
```

```python
forecast = results.forecast(horizon = 5)
forecasted_variance = forecast.variance.iloc[-1]

print(forecasted_variance)
```
